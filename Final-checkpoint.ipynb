 Data Preprocessing and Feature Engineering
 
This involves handling any remaining issues, encoding categorical variables, and scaling numerical features. We'll use a ColumnTransformer within a Pipeline for clean and reproducible preprocessing.

```python
# Divisons les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y_ok, test_size=0.30, random_state=42, stratify=y_ok)

# Identifions les variables selon leurs types
numerical_features = ['age', 'rest-bp', 'serum-chol', 'max-heart-rate', 'oldpeak']
categorical_features = ['sex', 'chest-pain', 'fasting-blood-sugar', 'electrocardiographic', 'angina', 'slope', 'major-vessels', 'thal']

# Créer un préprocesseur
# 1. Standard Scaler pour les caracteristiques numeriques
# 2. One-Hot Encoder pour les variables categorielles
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_features)
    ],
    remainder='passthrough'
)
```

Initial Modeling
  
We'll start with two common and robust classification algorithms: Logistic Regression (simple, interpretable baseline) and Random Forest (more powerful ensemble method).

```python
# Modelisation Initiale 

# Pipeline pour la Regression Logistique
logreg_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(solver='liblinear', random_state=42))
])

# Entrainement du modele
logreg_pipeline.fit(X_train, y_train)

# Prediction et Evaluation 
y_pred_logreg = logreg_pipeline.predict(X_test)
y_proba_logreg = logreg_pipeline.predict_proba(X_test)[:, 1]

print("\n Modele Initail : Performance de la Regression Logistique ")
print("Matrice de Confusion:")
print(confusion_matrix(y_test, y_pred_logreg))

print("\n Rapport de classification:")
print(classification_report(y_test, y_pred_logreg))
print(f"ROC AUC Score: {roc_auc_score(y_test, y_proba_logreg):.4f}")


# Pipeline pour le Random Forest
rf_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])

# Entrainement du modele
rf_pipeline.fit(X_train, y_train)

# Prediction et Evaluation
y_pred_rf = rf_pipeline.predict(X_test)
y_proba_rf = rf_pipeline.predict_proba(X_test)[:, 1]

print("\n Modele: Performance du Random Forest ")
print(" Matrice de Confusion:")
print(confusion_matrix(y_test, y_pred_rf))

print("\n Rapport de Classification:")
print(classification_report(y_test, y_pred_rf))
print(f"ROC AUC Score: {roc_auc_score(y_test, y_proba_rf):.4f}")
```

- Model Optimization (Hyperparameter Tuning)
  
We will use GridSearchCV to systematically search for the best combination of hyperparameters for the Random Forest first classifier and then Logistic Regression classifier, which in this case shows the best potential.

We know if we want to maximize the performance gain from tuning, it'll be preferable to choose the Random forest But 

```python
# Optimisation de Modele (Random Forest)

# Definissons les parametres a rechercher
param_grid = {
    'classifier__n_estimators': [50, 100, 200], 
    'classifier__max_depth': [None, 10, 20],   
    'classifier__min_samples_split': [2, 5],   
    'classifier__min_samples_leaf': [1, 2]     
}

# Configuration de GridSearchCV avec une validation croisee en 5 etapes
grid_search_rf = GridSearchCV(
    rf_pipeline,
    param_grid,
    cv=5,
    scoring='recall',
    verbose=1,
    n_jobs=-1 
)

# Recherche de la grille de depart 
grid_search_rf.fit(X_train, y_train)

# Evaluation du meilleur modele
best_rf_model = grid_search_rf.best_estimator_
y_pred_tuned_rf = best_rf_model.predict(X_test)
y_proba_tuned_rf = best_rf_model.predict_proba(X_test)[:, 1]

print("\n Modele Optimise : Performance du Random Forest (Optimise) ")
print(f"Best Parameters: {grid_search_rf.best_params_}")
print("Matrice de Confusion:")
print(confusion_matrix(y_test, y_pred_tuned_rf))
print("\n Rapport de Classification:")
print(classification_report(y_test, y_pred_tuned_rf))
print(f"ROC AUC Score: {roc_auc_score(y_test, y_proba_tuned_rf):.4f}")
```
